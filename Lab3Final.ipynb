```python
import sys
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
Book = open('romeoandjuliet.txt').read()
```

# File Functions

```python
def prepairfile(a):
    lst = (a.lower()).split()
    newlst = []
    for car in lst:
        for elm in car:
            if elm == '!' or elm == '?' or elm == ',' or elm == '.' or elm == ';' or elm == ':' or elm == '-' or elm == ')' or elm == '(' or elm == '[' or elm == ']':
                car = car.replace(elm,'')
        newlst += [car]
    return newlst

def word_finder(file):
    word_dict = {}
    for word in file:
        if word not in word_dict:
            word_dict[word] = 1
        elif word in word_dict:
            word_dict[word] += 1
    return word_dict

def key_order(word_dict):
    return sorted(word_dict.keys())

def value_order(word_dict):
    return sorted(word_dict.values())
```

**prepairfile:** This function removes all upper case letters and makes them lower case. This function also removes all punctuation charicters which are identified as different words. (ex: 'you,' and 'you!')

**word_finder:** This function puts all words in a dictionary and counts how many times they apear.

**key_order:** This function orders the dictionary created in word_finder by keys

**value_order:** This function orders the dictionary created in word_finder by values

# Removing punctuation Characters

```python
prepairfile(Book)
```

# Ordered dictionary by Values (decreasing)

```python
Dict = word_finder(prepairfile(Book))
orderD = value_order(Dict)
print(orderD)
```
